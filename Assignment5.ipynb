```python
import pandas as pd

def create_records():

    student_data = {
        "RollNo": [301, 302, 303, 304, 305],
        "StudentName": ["Kiran", "Sneha", "Vamsi", "Divya", "Arjun"],
        "Department": ["AI", "CSE", "ECE", "IT", "EEE"],
        "CGPA": [9.2, 8.4, 7.9, 8.8, 7.5],
        "Attendance": [96, 82, 78, 91, 85],
        "Age": [20, 21, 19, 22, 20],
        "Hosteller": [True, False, True, False, True]
    }

    df = pd.DataFrame(student_data)

    df["StudentName"] = df["StudentName"].astype("string")
    df["Department"] = df["Department"].astype("string")

    return df


def main():

    df = create_records()

    print("\nComplete DataFrame:\n")
    print(df)

    print("\nColumn Data Types:\n")
    print(df.dtypes)

    numeric_columns = df.select_dtypes(include='number')

    print("\nNumeric Data:\n")
    print(numeric_columns)

    string_columns = df.select_dtypes(include='string')

    print("\nString Data:\n")
    print(string_columns)

    boolean_columns = df.select_dtypes(include='bool')

    print("\nBoolean Data:\n")
    print(boolean_columns)

    non_numeric_columns = df.select_dtypes(exclude='number')

    print("\nNon Numeric Data:\n")
    print(non_numeric_columns)


if __name__ == "__main__":
    main()
```
